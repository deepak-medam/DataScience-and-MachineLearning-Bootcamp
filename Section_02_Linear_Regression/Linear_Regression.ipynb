```python
import matplotlib.pyplot as plt
import pandas
from pandas import DataFrame as pd
from sklearn.linear_model import LinearRegression
```

```python
cost_revenue = pandas.read_csv("/content/drive/MyDrive/2023-24/courses/udemy/Machine-Learning-Bootcamp/Resources/Section_02_Linear_Regression/cost_revenue_clean.csv")
```

```python
cost_revenue.describe()
```

```python
pd_budget = pd(cost_revenue, columns = ['production_budget_gpb']) # Production budget
ww_gross = pd(cost_revenue, columns=['worldwide_gross_gbp']) # reven
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pd_budget, ww_gross, alpha = 0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(pd_budget, ww_gross)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(pd_budget, ww_gross, alpha = 0.3)
plt.plot(pd_budget, regression.predict(pd_budget), color='red', linewidth=4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0, 3000000000)
plt.xlim(0, 450000000)
plt.show()
```
